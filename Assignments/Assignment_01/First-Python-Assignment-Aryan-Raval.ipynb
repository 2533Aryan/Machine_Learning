# Reading csv file 

```python
# This reads our data file and puts it in the variable records as a list of lists

with open('population-under-five-years-old.csv') as f:
    records = [row for row in csv.reader(f)]  # each sub-list is one row of the data
```

# Different countries/entity in our record

```python
# storing first column in variable country
country = [row[0] for row in records]
country = country[1:]  # remove attribute name "entity"

# Set function to get the unique entity
unique_country = set(country)
unique_country
```

```python
# number of countries
len(unique_country)
```

# Converting into float

```python
def try_float(s):
    try:
        f = float(s)
    except ValueError:
        return s
    return f
```

```python
type(try_float("1.4"))
```

```python
type(try_float("abc"))
```

```python
# convert numbers into float
records = [[try_float(x) for x in row] for row in records]
```

```python
records
```

# Year Range

```python
# storing third column in variable year
year = [row[2] for row in records]
year = year[1:]  # remove attribute name "year"

# Set function to get year range
unique_year = set(year)
unique_year
```

```python
# Total number of year
len(unique_year)
```

# Maximum population

```python
# storing fourth column in variable maximum_population
maximum_population = [row[3] for row in records]
maximum_population = maximum_population[1:]  # Remove attribute name "population"

maximum_population = max(maximum_population)
maximum_population
```

# Average (Median) population

```python
# storing fourth column in variable median_population
median_population = [row[3] for row in records]
median_population = median_population[1:]  # Remove attribute name "population"

median_population = statistics.median(median_population)
median_population
```

# Minimum population

```python
# storing fourth column in variable minimum_population
minimum_population = [row[3] for row in records]
minimum_population = minimum_population[1:]  # Remove attribute name "population"

minimum_population = min(minimum_population)
minimum_population
```

# Graph

- I am interested in findind out how the total population each year varies with time
- Hence, I end up plotting graph => total population during that year VS year

```python
# Store axes - x and y
graph_axes = []

# count total population
total_population = 0
for i in unique_year:
    for row in records:
        if row[2] == i:
            total_population = total_population + row[3]
    y,x = total_population,i
    graph_axes.append([x,y])  # storing them as list [x,y]

# Show list    
graph_axes
```

```python
# Plotting

x_axes = [row[0] for row in graph_axes] 
y_axes = [row[1] for row in graph_axes]

plt.plot(x_axes, y_axes)

plt.xlabel("Year")
plt.ylabel("Total Population")

plt.show()
```
